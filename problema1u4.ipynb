```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#Generar datos para la distribucion normal
media=70
desv_std=4
total_estudiantes=500
peso_limite=72
prob_menos_72kg=norm.cdf(peso_limite,media,desv_std)
estudiantes_menos_72kg=total_estudiantes*prob_menos_72kg
print("Probabilidad de que un estudiante pese menos de 72 kg:", prob_menos_72kg)
print("Estudiantes que pesan menos de 72 kg:", estudiantes_menos_72kg)
x=np.linspace(media - 4* desv_std, media + 4* desv_std, 1000)
y=norm.pdf(x,media,desv_std)
plt.plot(x,y, 'b',linewidth=2)
plt.fill_between(x, y, where=(x<= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 72 kg.\n' + r' $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
peso_limite=68
prob_menos_68kg=norm.cdf(peso_limite,media,desv_std)
prob_mas_68kg=1-prob_menos_68kg
estudiantes_mas_68kg=total_estudiantes*prob_mas_68kg
print("Probabilidad de que un estudiante pese mas de 68 kg:", prob_mas_68kg)
print("Estudiantes que pesan mas de 68 kg:", estudiantes_mas_68kg)
x=np.linspace(media - 4* desv_std, media + 4* desv_std, total_estudiantes)
y=norm.pdf(x,media,desv_std)
plt.plot(x,y, 'b',linewidth=2)
plt.fill_between(x, y, where=(x>= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
prob_entre=prob_menos_72kg-prob_menos_68kg
estudiantes_entre=total_estudiantes*prob_entre
print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", prob_entre)
print("Estudiantes que pesan entre 68 y 72 kg:", estudiantes_entre)
x=np.linspace(media - 4* desv_std, media + 4* desv_std, total_estudiantes)
y=norm.pdf(x,media,desv_std)
plt.plot(x,y, 'b',linewidth=2)
plt.fill_between(x, y, where=((x>= 68) & (x<=72)), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68 kg y menos de 72')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
