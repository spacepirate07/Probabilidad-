```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,10,20,50])
p=np.array([0.50,0.30,0.15,0.05])
print("Suma de probabilidades:",sum(p))
esperanza=np.sum(x*p)
esperanza_cuadrado=np.sum(x**2*p)
varianza=esperanza_cuadrado-esperanza**2
desviacion_estandar=np.sqrt(varianza)
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviacion Estandar = {desviacion_estandar:.2f}")
F=np.cumsum(p)
for xi,fi in zip(x, F):
  print(f"F({xi}) = {fi:.2f}")
plt.step(x, F, where='post')
plt.title("Funcion de distribucion acumulada")
plt.xlabel("Monto de premio")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0,1.05)
plt.show()
```

El valor esperado representa la ganancia promedio que obtiene un jugador a largo plazo. En este caso, indica que el jugador ganaría aproximadamente $8.50 por juego en promedio.

Considero que el juego no es tan favorable, ya que la probabilidad mas alta es de obtener 0 pesos (%50) y la probabilidad de sacar el premio mayor (50  pesos) es minima, de solo %5.
