```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3,4,5])
p=np.array([0.05,0.10,0.25,0.30,0.20,0.10])
print("Suma de probabilidades:",sum(p))
esperanza=np.sum(x*p)
esperanza_cuadrado=np.sum(x**2*p)
varianza=esperanza_cuadrado-esperanza**2
desviacion_estandar=np.sqrt(varianza)
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviacion Estandar = {desviacion_estandar:.2f}")
F=np.cumsum(p)
for xi,fi in zip(x, F):
  print(f"F({xi}) = {fi:.2f}")
plt.step(x, F, where='post')
plt.title("Funcion de distribucion acumulada")
plt.xlabel("Unidades vendidas")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0,1.05)
plt.show()
```

El valor esperado es aproximadamente 2.8, lo que indica que en promedio se venden cerca de 3 unidades por semana.
La desviación estándar es cercana a 1.25, lo que muestra que existe una variabilidad moderada en las ventas.

Esto significa que la mayoría de las semanas las ventas se encuentran alrededor del promedio, generalmente entre 2 y 4 unidades, por lo que la demanda no presenta cambios extremos.
La demanda es relativamente estable, ya que la variación no es muy grande y los valores se concentran cerca del promedio.
