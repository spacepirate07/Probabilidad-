```python
import math as mt
from itertools import permutations
asistentes=["Oscar","Paco","Paty","Laura","Ruben","Luis","Lucy","Alberto","Juan","Danny"]
n=len(asistentes)
r=4
Pr=mt.factorial(n)/(mt.factorial(n-r))
print("Existen ", Pr , " posibles permutaciones sin repeticion del total de",n, "nombres en grupos de", r ,"en ", r)
print("Aqui las permutaciones")
permutaciones_nombres=list(permutations(asistentes,r))
for permutacion in permutaciones_nombres:
  print(permutacion)
```
