#Modelo de regresion lineal

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import numpy as np
horas=df1['HorasEstudio'].values.reshape(-1,1)
calificacion=df1['Calificacion'].values
modelo=LinearRegression()
modelo.fit(horas, calificacion)
horas1=df1['HorasEstudio'].values
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)
coeficiente_correlacion, p_valor=pearsonr(horas1, calificacion)
print("Coeficiente de correlacion de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```

#Visualizacion de los resultados

```python
import seaborn as sns
plt.figure(figsize=(6,4))
horas=df1['HorasEstudio'].values
calificacion=df1['Calificacion'].values
sns.set(style="darkgrid")
sns.regplot(x=horas, y=calificacion, scatter=True, color='blue')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacion')
plt.title('Diagrama de dispersion con linea de regresion')
plt.grid(True)
plt.show()
```

#Calculo de la correlacion

```python
horas1=df1['HorasEstudio'].values
coeficiente_correlacion, p_valor=pearsonr(horas1, calificacion)
print("Coeficiente de correlacion de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```

##Preguntas grupo 1
###1. ¿Cuál es la interpretación del coeficiente de regresión (pendiente)?
El coeficiente de regresión (m = -0.106) indica cuánto cambia la calificación por cada unidad que aumenta la variable independiente (horas). En este caso, la interpretación es que por cada hora adicional de estudio, la calificación final disminuye en promedio 0.106 puntos. Es una relación inversa, aunque el impacto es muy pequeño.
###2. ¿Cómo se interpreta el coeficiente de correlación en términos de fuerza y dirección?
Dirección: Es negativa, ya que el valor es menor a cero (-0.057). Esto significa que las variables se mueven en direcciones opuestas (a más horas, menos nota). Fuerza: Es extremadamente débil o nula. Al estar el valor tan cerca de 0, indica que no existe una relación lineal real entre el tiempo de estudio y las calificaciones. Los puntos en el diagrama de dispersión están muy alejados de la línea de regresión.
###3. ¿El modelo de regresión lineal es estadísticamente significativo?
No, no es estadísticamente significativo. Para que un modelo sea significativo, el valor p (p-value) debe ser menor a 0.05. En este caso, el valor p es de 0.693, lo cual es mucho mayor al límite. Esto significa que los resultados obtenidos son probablemente fruto del azar y no existe evidencia suficiente para afirmar que las horas de estudio influyen en la calificación.
###4. ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones
El modelo tiene una confiabilidad muy baja (nula) por tres razones: La correlación es casi cero. El valor p indica que no hay significancia estadística. El diagrama de dispersión muestra que los datos no siguen el patrón de la línea. Conclusión: Usar este modelo para predecir la nota de un alumno basándose en sus horas de estudio sería equivalente a una adivinanza; el error de predicción sería muy alto.

##Conclusion general del grupo 1
Tras realizar el análisis estadístico y la modelación lineal del Grupo 1, se concluye que no existe una relación funcional ni estadística entre el tiempo de estudio y las calificaciones finales de los estudiantes. Esta conclusión se sustenta en los siguientes puntos clave: Ausencia de Predictibilidad: El modelo de regresión lineal arrojó una pendiente casi nula (-0.11) y una intercepción de 85.31. Esto sugiere que, independientemente de las horas dedicadas, las calificaciones tienden a mantenerse en un rango similar, sin que el estudio adicional represente una mejora en el rendimiento. Falta de Significancia: El valor p de 0.693 es el indicador más contundente; al superar ampliamente el umbral de 0.05, nos indica que cualquier ligera tendencia observada es puramente accidental y no tiene validez científica para este conjunto de datos. Correlación Nula: El coeficiente de Pearson (-0.057) confirma que la fuerza de asociación entre las variables es inexistente. En el diagrama de dispersión, esto se traduce en puntos dispersos aleatoriamente que no se alinean con la trayectoria de la recta. Veredicto Final: El tiempo de estudio no es un factor determinante para el éxito académico en este grupo específico.
