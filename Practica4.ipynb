##9. Diagrama de cajas y bigotes
1. La lineas se usan para indicar la variabilidad fuera de los cuartiles superior e inferior.
2. Los valores atipicios se representan a veces como puntos individuales que estan en linea con los bigotes.
3. **Si los datos son simetricos, la mediana estara en el centro de la caja.**
4. Si los datos estan **sesgados**, la mediana estara mas cerca de la parte superior o inferior de la caja.
5. Los extremos de arriba y abajo de la caja indican los **cuantiles, o percentiles, 25 y 75.** Estos dos cuantiles tambien se conocen como cuartiles, porque separando cuartos (25%) de los datos. La  longitud de la caja es la diferencia entre estos dos percentiles y se conoce como rango intercurtilico (IQR).
6. Los diagramas caja-bigotes son muy utiles para comparar dos o mas conjuntos de datos.

```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(40, 93), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
plt.boxplot(fabrica1)
plt.ylabel('Produccion en mts.')
plt.title('Produccion de la fabrica 1')
plt.show()
```

```python
plt.boxplot(fabrica2)
plt.ylabel('Produccion en mts.')
plt.title('Produccion de la fabrica 2')
plt.show()
```

##**Comparamos la produccion 1 y 2 en un subplot**

```python
fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1=ax1.boxplot(fabrica1,
                  vert=True, #Alineacion vertical
                  patch_artist=True) #Rellenar con color
ax1.set_title('Produccion fabrica 1')
bplot2=ax2.boxplot(fabrica2,
                  vert=True, #Alineacion vertical
                  patch_artist=True) #Rellenar con color
ax2.set_title('Produccion fabrica 2')
plt.show()
```

```python
#Insertamos un valor de 150 a la fabrica 1, seria un valor atipico
fabrica1.append(150)
print(fabrica1)
```

```python
fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1=ax1.boxplot(fabrica1,
                  vert=True, #vertical box alignment
                  patch_artist=True)
ax1.set_title('Produccion fabrica 1')
bplot2=ax2.boxplot(fabrica2,
                  vert=True, #vertical box alignment
                  patch_artist=True)
ax2.set_title('Produccion fabrica 2')
plt.show()
```

```python
#Agregamos ax1.set_ylim(0, 160) para garantizar que ambos graficos usen la misma escala virtual, facilitando la comparacion visual.
fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, figsize=(9,5))
bplot1=ax1.boxplot(fabrica1,
                   vert=True, #vertical box alignment
                  patch_artist=True) #Relenar con color
ax1.set_title('Produccion fabrica 1')
ax1.set_ylim(30, 160) #Escala uninforme de 0 a 160
bplot2=ax2.boxplot(fabrica2,
                   vert=True, #Alineacion vertical t
                  patch_artist=True) #Rellenar con color
ax2.set_title('Produccion fabrica 2')
ax2.set_ylim(30, 160) #Escala uniforme de 0 a 160
plt.show()
```
