```python
n=len(autos.price)
k=int(1+math.log2(n))
print("Valor de n = ", n ,"\nNumero de intervalos =", k)
```

```python
plt.hist(autos.price, bins=k, alpha=0.5)
plt.title('Histograma de precios basado en la regla de sturges')
plt.xlabel('Precios')
plt.ylabel('Frecuencia')
```

El histograma muestra que la mayoria de precios del archivo ronda entre los 5000 y los 10000 y poco a poco la frecuencia de los precios mayores a 10000 va descendciendo gradualmente hasta el maximo que son los 45000, el cual es minimo, aunque el minimo esta entre los 25000 y 30000.

```python
sns.boxplot(x="price", y ="body-style", data=autos)
plt.xlabel('Precios')
plt.ylabel('Tipo de vehiculo')
```

El diagrama muestra la distribuciÃ³n de los precios segÃºn el tipo de vehÃ­culo. Se observa que:

-Hardtop y convertible tienen las medianas mÃ¡s altas, por lo que suelen ser los mÃ¡s caros.

-Hatchback presenta la mediana mÃ¡s baja, siendo el tipo mÃ¡s econÃ³mico.

-Sedan y wagon se ubican en un rango intermedio.

-Hardtop y convertible muestran mayor dispersiÃ³n, lo que indica mÃ¡s variabilidad en precios.

-Sedan presenta varios valores atÃ­picos altos, lo que representa que hay modelos que superan por mucho el rango de precio tipico.

```python
sns.boxplot(x="price", y ="drive-wheels", data=autos)
plt.xlabel('Precios')
plt.ylabel('Tipo de traccion ruedas')
```

-rwd (tracciÃ³n trasera) presenta la mediana mÃ¡s alta y la mayor dispersiÃ³n de precios. AdemÃ¡s, tiene varios valores atÃ­picos altos, lo que indica que incluye vehÃ­culos mÃ¡s costosos (posiblemente deportivos o de gama alta).

-fwd (tracciÃ³n delantera) tiene una mediana mÃ¡s baja y precios mÃ¡s concentrados en rangos econÃ³micos y medios. Presenta algunos valores atÃ­picos, pero en general es el grupo mÃ¡s accesible.

-4wd (tracciÃ³n en las cuatro ruedas) muestra precios intermedios, con menor variabilidad que rwd. Tiene algÃºn valor atÃ­pico, pero su rango es mÃ¡s compacto.
