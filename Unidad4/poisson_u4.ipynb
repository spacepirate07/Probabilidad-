```python
from scipy.stats import poisson
media=5;
probabilidad_2_personas=poisson.pmf(2,media);
print(probabilidad_2_personas);
print("a) Probabilidad de que haya 2 personas ingresadas a cuidados intensivos: ", probabilidad_2_personas)
```

```python
media=5
probabilidad_10_o_mas=1-poisson.cdf(10,media);
print(probabilidad_10_o_mas);
print("b) Probabilidad de que haya 10 o mas: ", probabilidad_10_o_mas)
```

  Es probable que X exceda de 10?
  Es posible (ya que la probabilidad es mayor que 0) pero es muy poco probable, ya que la probabilidad es muy baja, de tan solo 1.3%.
