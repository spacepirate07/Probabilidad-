La solucion es binomial, ya que aqui si hay un numero definido (n) y tienen solo 2 resultados, defectuoso o no defectuoso (fracaso o exito)

```python
from scipy.stats import binom
probabilidad_fracaso=0.001
num_motores=1000
X=0
prob_ningun_defectuoso=binom.pmf(X, num_motores, probabilidad_fracaso)
print("La probabilidad de que ningun motor salga defectuoso es: ", prob_ningun_defectuoso)
```
